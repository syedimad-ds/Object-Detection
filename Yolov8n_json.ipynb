```python
from ultralytics import YOLO

# Load the nano model
model = YOLO('yolov8n.pt')

# EXPORT COMMAND: Format tfjs, aur imgsz 640 set kar rahe hain.
# Half=True se model chota aur fast ho jayega.
model.export(format='tfjs', imgsz=640, half=True)
```

```python
import shutil
shutil.make_archive('yolov8n_web_model', 'zip', 'yolov8n_web_model')
```
